## <font color='blue'> <b>array() </b><font color='white'>

- **NumPy**, bilimsel ve sayısal hesaplamalar için dizilerle -arrays- çalışmayı optimize eden bir Python kütüphanesidir.
- Bu kütüphanede yer alan çok boyutlu dizilere **ndarray** adı verilir. Bu isim, 'n-boyutlu dizi' anlamına gelir.
- **array()** fonksiyonunu kullanarak bir NumPy ndarray nesnesi oluşturabiliriz.

**np.array()** fonksiyonu, Python listelerini, tuple'ları ya da diğer dizi benzeri veri yapılarını NumPy dizisine (ndarray) dönüştürmek için kullanılır. 

**En sık kullanılan parametresi:**

**dtype**:
   - Dizide saklanacak öğelerin veri tipini belirtir.
   - Örnek: float, int, complex. Eğer belirtilmezse, NumPy veri tipini girdi veriye göre otomatik olarak belirler.

```python
#3darrayler genelde data visualizatiın ya da deep learningde karşımıza çıkıyor.
```

```python
my_list = [1, 2, 3]
my_list
```

```python
np.array(my_list)
```

```python
np.array([4, 5, 6.5])
```

```python
np.array([7, 8, 9], dtype=float)
```

```python
np.array([7, 8, 9], dtype=object)
```

```python
np.array([[1, 2], 
          [3, 4]])
```

```python
my_list2 =[[1, 2, 4], [3, 5, 6], [7, 9, 8], [10, 11, 12]] #herhangi bir row tek ya da 2 elemanlı olsaydı array yapamayıp hata verecekti.eksik eleman olmaz.fazla eleman da olmaz.
my_list2
```

```python
np.array(my_list2)
```

## <font color='blue'> <b>shape & reshape() </b><font color='white'>
**shape:** 
- Numpy array boyutunu yani kaç satır ve sütundan oluştuğunu gösteren bir tuple döndürür.
- Bu tuple'daki sayılar, dizinin her bir boyutundaki öğe sayısını gösterir. 
- Örneğin, (m, n) tuple'ı m satır ve n sütundan oluşan bir matrisi temsil eder.

**reshape():** 
- Numpy array’i yeniden şekillendirmek istediğimizde yani satır ve sütun sayısını değiştirmek istediğimizde reshape() metodunu kullanabiliriz.
- Diziyi belirtilen yeni şekle getirerek yeni bir ndarray döndürür. Ancak, bu metod orijinal veriyi değiştirmez.
- Belirli boyutlar için "-1" değeri kullanılabilir, bu sayede NumPy otomatik olarak belirtilen diğer boyutlara uygun şekilde bu boyutu hesaplar.
​

```python
my_arr = np.array(my_list2)
my_arr
```

```python
my_arr.shape() #shape bir attributedur fonk. değil bu nedenle parantez aç kapa yapılmaz.
```

```python
my_arr.shape #shape bir attributedur fonk. değil bu nedenle parantez aç kapa yapılmaz.
```

```python
np.array([1, 2, 3, 4, 5]).shape
```

```python
my_arr
```

```python
my_arr.reshape(3,4) #verileri sırayla alıp 4er adet satırlara dağıtarak reshape işlemini yapıyor. Yeniden bdüzenliyor, boyutlandırıyor.yeterki 12 elemanı boşta bırakmayacak şekilde bir düzen iste.her türlü şekillendirir.
```

```python
my_arr.reshape(2,6)
```

```python
my_arr.reshape(3,3)
```

```python
my_arr.reshape(3,5)
```

```python
my_arr
```

```python
my_arr.reshape(6,-1) #satır sayım 6 olsun da sütun sayım kaç olacaksa olsun bilmiyorum demek.
```

```python
my_arr.reshape(-1, 4) #sütun sayım 4 olsun da satır sayım kaç olacaksa olsun bilmiyorum demek.
```

```python
my_arr.reshape(-1, 5) #hata almamak için yine tam bölünme lazım.reshape sadece düzenler gösterir. my_array üzerinde kalıcı bir değişime sebep olmaz.
```

```python
my_arr.reshape(-1, -1) #en az brinin belirtilmesi lazım satır ya da sütun. yoksa hata verir
```
