```python
import numpy as np
import pandas as pd
```

# <font color='green'> <b>Missing Values</b><font color='black'>

```python
id_no = ["P001", "P002", "P003", "P004", "P005", "P006", "P007", "P008", "P009", "P010", "P011"]
gender = ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F", "M"]
status = ["FT", "PT", "-", "FT", "PT", "PT", "FT", "-", "PT", "FT", np.nan]
dept = ["DS", "FS", "AWS", "AWS", "DS", None, "FS", "FS", np.nan, "DS", "AWS"]
V1 = np.array([2, 3, 5, np.nan, 7, 1, np.nan, 10, 14, "-", 6])
V2 = np.array([8, np.nan, 5, 8, 11, np.nan, np.nan, 2, 3, 7, 9])
salary = np.array([np.nan, 54, 59, 120, 58, 75, None, 136, 60, 125, np.nan])

df0 = pd.DataFrame({
                   "id" : id_no,
                   "gender": gender,
                   "status": status,
                   "dept": dept,
                   "var1" : V1,
                   "var2" : V2,
                   "salary" : salary
                  })
```

```python
df = df0.copy()       # df0 ın orjinal olarak kalmasını istiyoruz
df
```

```python
df["var1"] = df["var1"].replace(to_replace = "-", value = np.nan).astype("float")
df["var1"]
```

```python
df["status"] = df["status"].replace("-", np.nan)
df["status"]
```

```python
df.isnull().sum()
```

```python
df.groupby("status")["salary"].mean() #buradan salary 90 altıysa statusu part time ile üstüyse  fulltime olarak kabul et diyebiliriz.
```

```python
df.loc[df.salary > 90, "status"] #salarysi 90dan büyük olan satırların(verilerin) status değerini getir.
```

```python
df.loc[df.salary <= 90, "status"] #salarysi 90dan küçük veya eşit satırların(verilerin) status değerini getir.
```

```python
df.loc[df.salary <= 90, "status"].mode()[0]
```

```python
df.loc[df.salary > 90, "status" ].fillna(df.loc[df.salary > 90, "status" ].mode()[0], inplace = True) #bazen birden fazla koşul durumu olduğunda inplace contaları yakıyor, çalışmıyor.burdaki gibi
```

```python
df
```

```python
df.loc[df.salary > 90, "status" ]= df.loc[df.salary>90,"status"].mode()[0] #yukardaki koddaki gibi inplace çalışmazsa fillna yerine atama yaparız
df
```

```python
df.loc[df.salary <= 90, "status" ]= df.loc[df.salary <= 90,"status"].mode()[0] 
df #statuste son bir nan değer kaldı.çünkü o satırda salary de nan olduğu iiçin kodumun filtresine girmedi.
```

```python
#status sütunundaki son eksik değeri "gender" ve "dept" grubunun modu ile dolduralım.

df.groupby(["gender", "dept"])["status"].transform(lambda x : x.mode()[0]) 

#bunu yaptığımda tüm değerlere bunu yaptı ama ben nan değere bu işlemi yapmak istiyorum
#bu ve alttaki satırı aşama aşama incele karıışık biraz.kodu parçalayip incele.
```

```python
df.groupby(["gender", "dept"])["status"].mode()#neden çalışmadı?
```

```python
df.groupby(["gender", "dept"])["status"].value_counts()#yukardaki çalışmadığı için bunu görmek istedim.
```

```python
df.status.fillna(df.groupby(["gender", "dept"])["status"].transform(lambda x : x.mode()[0]))#artık 3 yukardaki işlemi nan satırda yapıyor.
```

```python
df.status.fillna(df.groupby(["gender", "dept"])["status"].transform(lambda x : x.mode()[0]), inplace=True)#artık 3 yukardaki işlemi nan satırda yapıyor.
```
